### **Question 13:** The company wants to provide a promotional offer in the advertisement of the restaurants. The condition to get the offer is that the restaurants must have a rating count of more than 50 and the average rating should be greater than 4. Find the restaurants fulfilling the criteria to get the promotional offer. [3 marks]

```python
# Write the code here
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df_foodhub_order = pd.read_csv('foodhub_order.csv')

# Count the number of rows for each `restaurant_name` where the value in the `rating` column is not 'Not given'
rating_counts = df_foodhub_order[df_foodhub_order['rating'] != 'Not given']['restaurant_name'].value_counts()

# Filter the data to include only the restaurants with a rating count of more than 50
df_filtered = df_foodhub_order[df_foodhub_order['restaurant_name'].isin(rating_counts[rating_counts > 50].index)]

# Convert the `rating` column to numeric
df_filtered['rating'] = pd.to_numeric(df_filtered['rating'], errors='coerce')

# Calculate the average of the `rating` column for each `restaurant_name`
average_ratings = df_filtered.groupby('restaurant_name')['rating'].mean()

# Filter the data to include only the restaurants with an average rating of more than 4
df_filtered = df_filtered[df_filtered['restaurant_name'].isin(average_ratings[average_ratings > 4].index)]

# Display the `restaurant_name` for the filtered data
print(f'The restaurants fulfilling the criteria to get the promotional offer are:\n{df_filtered["restaurant_name"].unique()}')
```

### **Question 14:** The company charges the restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders. [3 marks]

```python
# Write the code here
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df_foodhub_order = pd.read_csv('foodhub_order.csv')

# Calculate the revenue for orders above $20
revenue_above_20 = df_foodhub_order[df_foodhub_order['cost_of_the_order'] > 20]['cost_of_the_order'] * 0.25
# Calculate the revenue for orders between $5 and $20
revenue_above_5 = df_foodhub_order[(df_foodhub_order['cost_of_the_order'] > 5) & (df_foodhub_order['cost_of_the_order'] <= 20)]['cost_of_the_order'] * 0.15

# Calculate the total revenue
total_revenue = revenue_above_20.sum() + revenue_above_5.sum()

# Print the total revenue
print(f'The net revenue generated by the company across all orders is: ${total_revenue.round(2)}')
```

### **Question 15:** The company wants to analyze the total time required to deliver the food. What percentage of orders take more than 60 minutes to get delivered from the time the order is placed? (The food has to be prepared and then delivered.) [2 marks]

```python
# Write the code here
# Write the code here
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df_foodhub_order = pd.read_csv('foodhub_order.csv')

# Calculate the total delivery time
df_foodhub_order['total_time'] = df_foodhub_order['food_preparation_time'] + df_foodhub_order['delivery_time']

# Calculate the percentage of orders having more than 60 minutes of total delivery time
percentage = (len(df_foodhub_order[df_foodhub_order['total_time'] > 60]) / len(df_foodhub_order)) * 100

# Print the percentage
# Use the round() function instead of percentage.round()
print(f'{round(percentage, 2)}% of orders take more than 60 minutes to get delivered from the time the order is placed.')
```

### **Question 16:** The company wants to analyze the delivery time of the orders on weekdays and weekends. How does the mean delivery time vary during weekdays and weekends? [2 marks]

```python
# Write the code here
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Read the CSV file into a DataFrame
df_foodhub_order = pd.read_csv('foodhub_order.csv')

# Calculate the mean delivery time for weekdays
mean_delivery_time_weekdays = df_foodhub_order[df_foodhub_order['day_of_the_week'] == 'Weekday']['delivery_time'].mean()
# Calculate the mean delivery time for weekends
mean_delivery_time_weekends = df_foodhub_order[df_foodhub_order['day_of_the_week'] == 'Weekend']['delivery_time'].mean()

# Print the mean delivery time for weekdays and weekends
print(f'The mean delivery time on weekdays is: {mean_delivery_time_weekdays.round(2)}')
print(f'The mean delivery time on weekends is: {mean_delivery_time_weekends.round(2)}')
```
